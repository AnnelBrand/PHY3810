```python
# PHY3810
# Selecting the Data for Modeling
import pandas as pd

gammaRaySimulation_file_path = '/Users/Kunwiji/Dropbox/AnnelB/run_020963_DST_001.root_CT1.csv'
gammaRaySimulation_data = pd.read_csv(gammaRaySimulation_file_path) 
gammaRaySimulation_data.columns
```

```python
# Selecting The Prediction Target
y = gammaRaySimulation_data.HillasParameters_Hillas0510_1_Known
```

```python
# Choosing "Features"
gammaRaySimulation_features = ['HillasParameters_Hillas0510_1_NPixels', 
                               'HillasParameters_Hillas0510_1_NPixels', 
                               'HillasParameters_Hillas0510_1_Length', 
                               'HillasParameters_Hillas0510_1_Length', 
                               'HillasParameters_Hillas0510_1_Skewness', 
                               'HillasParameters_Hillas0510_1_Kurtosis', 
                               'HillasParameters_Hillas0510_1_CoG_x', 
                               'HillasParameters_Hillas0510_1_CoG_y', 
                               'HillasParameters_Hillas0510_1_Phi']
```

```python
X = gammaRaySimulation_data[gammaRaySimulation_features]
```

```python
# Reviewing the data by using the describe method and the head method, which shows the top 
#few rows.
X.head()
```

```python
X.describe()
```

```python
# Building the Model
from sklearn.tree import DecisionTreeClassifier
```

```python
# Define model. Specify a number for random_state to ensure same results each run.
gammaRaySimulation_model = DecisionTreeClassifier(random_state=1)
```

```python
# Fit model
gammaRaySimulation_model.fit(X, y)
```

```python
# Make predictions for the first few rows of the training data to see how the predict function works.
print("Making predictions for the first 5 rows:")
print(X.head())
print("The predictions are")
print(gammaRaySimulation_model.predict(X.head()))
```
